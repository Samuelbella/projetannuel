```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the training data
train_df = pd.read_csv('train.csv')
train_df
```

### 1. Afficher une image
Afficher une image quelconque depuis le train en indiquant comme titre de l'image son
label (càd le target associé) ?

```python
def display_digit(index, data, labels):
    label = labels[index]
    image = data.iloc[index].values.reshape(28, 28)
    plt.imshow(image, cmap='gray')
    plt.title(f'Label: {label}')
    plt.axis('off')
    plt.show()

random_index = np.random.randint(0, len(train_df))
display_digit(random_index, train_df.drop('label', axis=1), train_df['label'])
```
